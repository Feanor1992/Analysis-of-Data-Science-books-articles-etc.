```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
```

```python
df = pd.read_csv('nhanes_2015_2016.csv')
```

```python
df
```

* RIAGENDR - gender
* DMDMARTL - marital status
* DMDEDUC2 - level of education
* BPXDI1 - distolic blood pressure
* BPXSY1 - systolic blood pressure
* BPXSY1 - normalize systolic blood pressure
* RIDAGEYR - age
* BMXBMI - BMI (body mass index)
* RIDRETH1 - ethnicity
* BMXHT - height
* BMXWT - weight
* DMDHHSIZ - house size

```python
df['RIAGENDRx'] = df.RIAGENDR.replace({1: "Male", 2: "Female"})
df['DMDEDUC2x'] = df.DMDEDUC2.replace({1: "<9", 2: "9-11", 3: "HS/GED", 4: "Some college/AA", 5: "College", 7: "Refused", 9: "Don't know"})
df['DMDMARTLx'] = df.DMDMARTL.replace({1: "Married", 2: "Widowed", 3: "Divorced", 4: "Separated", 5: "Never married", 6: "Living w/partner", 77: "Refused"})
df
```

# Scatterplot relations between distolic blood pressure and systolic blood pressure


```python
# category of unique martial status
category_of_martial_status = df['DMDMARTLx'].unique()
category_of_martial_status
```

```python
# choosing color
#colors = [plt.cm.tab10(i/float(len(category_of_martial_status)-1)) for i in range(len(category_of_martial_status))]

# visualizing the scatterplot
plt.figure(figsize=(18, 12), dpi=80, facecolor="w", edgecolor="k")

for i, category in enumerate(category_of_martial_status):
    plt.scatter(x='BPXDI1', y='BPXSY1', data=df.loc[df.DMDMARTLx == category, :], s=20, label=str(category))
    
plt.gca().set(xlabel='BPXDI1', ylabel='BPXSY1')
    
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title('Relation betwqeen Marital status, Systolic blood pressure and Distolic blood pressure', fontsize=18)
plt.legend(fontsize=12)
plt.show()
```

Select first 500 rows and person who the older than 40 years old

```python
from scipy.spatial import ConvexHull

df_first_500 = df.loc[:499, :]
df['dot_size'] = df.BMXBMI*10
df_first_500
```

```python
 # select person who elsder than 40 yo
df_40_yo = df_first_500.loc[(df_first_500["RIDAGEYR"] > 40), :].dropna()
df_40_yo
```

```python
fig = plt.figure(figsize=(24, 12), dpi= 80, facecolor='w', edgecolor='k')

for i, category in enumerate(category_of_martial_status):
    plt.scatter(x='BPXDI1', y='BPXSY1', data=df_first_500.loc[df_first_500.DMDMARTLx == category, :], s='dot_size', label=str(category))
    
def encircle(x,y, ax=None, **kw):
    if not ax: 
        ax=plt.gca()
    
    p = np.c_[x,y]
    hull = ConvexHull(p)
    poly = plt.Polygon(p[hull.vertices,:], **kw)
    ax.add_patch(poly)

# Drawing a polygon surrounding vertices    
encircle(df_40_yo.BPXDI1, df_40_yo.BPXSY1, ec="k", fc="gold", alpha=0.1)
encircle(df_40_yo.BPXDI1, df_40_yo.BPXSY1, ec="firebrick", fc="none", linewidth=1.5)

plt.gca().set(xlabel='BPXDI1', ylabel='BPXSY1')
plt.xticks(fontsize=12); plt.yticks(fontsize=12)
plt.title("Relation betwqeen Marital status, Systolic blood pressure and Distolic blood pressure from 40+ yo persons.", fontsize=22)
plt.legend(fontsize=12)    
plt.show()
```

inside the rectangle there person who elder than 40 yo. 
The size of the circles corresponds to the BMI - the larger the circle, the higher the index.

**There is no relationship between blood pressure and martial status**

## diverging bars

Usin **diverging bars** helps to estimate how much data deviates from one metric.

```python
# relation between normalize systolic blood pressure and house size
x = df_first_500.loc[:, "BPXSY1"]

df_first_500["BPXSY1_n"] = (x - x.mean())/x.std()

# Red marks negative values, blue marks positive values.
df_first_500['colors'] = ['red' if i < 0 else 'blue' for i in df_first_500["BPXSY1_n"]]
df_first_500.sort_values("BPXSY1_n", inplace=True)
df_first_500.reset_index(inplace=True)

plt.figure(figsize=(16, 10), dpi=80)
plt.hlines(y = df_first_500.DMDHHSIZ, xmin=0, xmax = df_first_500.BPXSY1_n, color=df_first_500.colors, linewidth=3)
plt.gca().set(ylabel="DMDHHSIZ", xlabel = "BPXSY1_n")
plt.yticks(df_first_500.DMDHHSIZ, fontsize=14)
plt.grid(linestyle='--', alpha=0.5)
plt.show()
```
